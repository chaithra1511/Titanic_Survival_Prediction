### Step 4: Missing Values Treatment

##### We can observe that there are 177 null values in Age column.Since, Age is Skewed,we replace the null values with median of all the ages.

```python
titanic_df['Age']=titanic_df['Age'].fillna(value=titanic_df['Age'].median())
```

```python
titanic_df['Embarked']=titanic_df['Embarked'].fillna(value='S')
```

### Step 5: Outliers

##### From Univariate analysis,we can conclude that from Age,Fare,SibSp and Parch Outliers can be detected.

#### ->Outlier Handling for Age:

```python
IQRage = titanic_df['Age'].quantile(0.75)-titanic_df['Age'].quantile(0.25)
```

```python
Upper_age = titanic_df['Age'].quantile(0.75) + 1.5*IQRage
Lower_age = titanic_df['Age'].quantile(0.25) - 1.5*IQRage
```

```python
Outlierage = titanic_df[(titanic_df['Age']>=Upper_age) | (titanic_df['Age']<=Lower_age)]
```

```python
l1=list(Outlierage.index)
for i in range(len(l1)):
    titanic_df['Age'][l1[i]] = titanic_df['Age'].quantile(0.95)
```

```python
sns.boxplot(y='Age', data=titanic_df)
```

#### ->Outlier handling for Fare:

```python
IQRfare = titanic_df['Fare'].quantile(0.75)-titanic_df['Fare'].quantile(0.25)
```

```python
Upper_fare = titanic_df['Fare'].quantile(0.75) + 1.5*IQRfare
Lower_fare = titanic_df['Fare'].quantile(0.25) - 1.5*IQRfare
print(Upper_fare)
```

```python
Outlierfare = titanic_df[(titanic_df['Fare']>=Upper_fare) | (titanic_df['Fare']<=Lower_fare)]
```

```python
l2=list(Outlierfare.index)
for i in range(len(l2)):
    titanic_df['Fare'][l2[i]] = titanic_df['Fare'].quantile(0.87)
```

```python
sns.boxplot(y='Fare', data=titanic_df)
```

#### ->Outlier handling for SibSp:

```python
IQRsib = titanic_df['SibSp'].quantile(0.75)-titanic_df['SibSp'].quantile(0.25)
```

```python
Upper_sib = titanic_df['SibSp'].quantile(0.75) + 1.5*IQRsib
Lower_sib = titanic_df['SibSp'].quantile(0.25) - 1.5*IQRsib
```

```python
Outliersib = titanic_df[(titanic_df['SibSp']>=Upper_sib) | (titanic_df['SibSp']<=Lower_sib)]
```

```python
l3=list(Outliersib.index)
for i in range(len(l3)):
    titanic_df['SibSp'][l3[i]] = titanic_df['SibSp'].quantile(0.93)
```

```python
sns.boxplot(y='SibSp', data=titanic_df)
```

#### ->Outlier handling for Parch

```python
IQRpar = titanic_df['Parch'].quantile(0.75)-titanic_df['Parch'].quantile(0.25)
```

```python
Upper_par = titanic_df['Parch'].quantile(0.75) + 1.5*IQRpar
Lower_par = titanic_df['Parch'].quantile(0.25) - 1.5*IQRpar
```

```python
Outlierpar = titanic_df[(titanic_df['Parch']>=Upper_par) | (titanic_df['Parch']<=Lower_par)]
```

### CLASSIFICATION

### Defining X and Y

```python
X=titanic_new.iloc[:,1:].values
```

```python
y=titanic_new.iloc[:,0].values
```

### Splitting the dataset

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25 , random_state=0)
```

### Building the model

### 1.Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
# Representing LinearRegression as lr(Creating LinearRegression Object)
lr = LogisticRegression()

# fit the model to the training data
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
y_test
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,f1_score,precision_score,recall_score,roc_auc_score
accuracy = accuracy_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_pred)

print('Accuracy is  :' ,round(accuracy,2)*100)
print('F1 score is :' ,round(f1,2)*100)
print('Precision is  :',round(precision,2)*100)
print('Recall is  :',round(recall,4)*100)
print('Roc Auc is  :',round(roc_auc,2)*100)
```
